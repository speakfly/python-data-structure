```python
# Queue() 创建一个空的新队列
# enqueue(item) 将新项添加到队尾  
# dequeue() 从队首移除项          
# isEmpty() 查看队列是否为空
# size()    返回队列中的项数

class Queue:
    def __init__(self):
        self.items = []
        
    def enqueue(self, item):
        self.items.insert(0, item)
        
    def dequeue(self):
        return self.items.pop()
    
    def isEmpty(self):
        return self.items ==[]
    
    def size(self):
        return len(self.items)
    
    
q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
print(q.size())
print(q.isEmpty())
print(q.dequeue())
print(q.size())
```

```python
# 模拟一个游戏
# 一群人围成一个圈， 指定之中一人为第一个人， 第7个人淘汰，从下一个开始重新计数

# 队列模拟这个圈
# 循环num次
# 安全的人 先出队后进队
# 循环完 淘汰出局的人

def hotPotato(namelist, num):
    simqueue = Queue()
    for name in namelist:
        simqueue.enqueue(name)
        
    while simqueue.size() > 1:
        for i in range(num):
            simqueue.enqueue(simqueue.dequeue())
            
        simqueue.dequeue()
        
    return simqueue.dequeue()

print(hotPotato(["Bill", "David", "Susan", "Jane", "Kent", "Brad"], 7))
print(hotPotato(["Bill", "David", "Susan"], 7))
```

```python
# 模拟打印机
#  需要创建三个类   Printer， Task, PrintQueue


# Printer 类  打印机
# tick()  模拟工作时， 时间一秒一秒流逝
# busy()  判断是否忙碌
# startNext(newtask)

class Printer:
    def __init__(self, ppm):
        self.pagerate = ppm          # 打印的速率
        self.currentTask = None
        self.timeRemaining = 0       # 需要工作的时间
    
    def tick(self):
        if self.currentTask != None:
            self.timeRemaining = self.timeRemaining - 1
            if self.timeRemaining <= 0:
                self.currentTask = None
    
    def busy(self):
        """判断是否忙碌"""
        if self.currentTask != None:
            return True
        else:
            return False
    
    def startNext(self, newtask):
        """接收任务 并计算任务时间"""
        self.currentTask = newtask
        self.timeRemaining = newtask.getPages() * 60/self.pagerate
        
        
# Task类 代表单个打印任务
# timestamp 表示  任务创建的时刻
# currenttime 表示 处理任务的时刻
import random

class Task:
    def __init__(self, time):  
        self.timestamp = time
        self.pages = random.randrange(1, 21)
    
    def getStamp(self):
        return self.timestamp
    
    def getPages(self):
        return self.pages
    
    def waitTime(self, currenttime):
        """返回等待的时间"""
        return currenttime - self.timestamp
    
    
# PrintQueue 对象 是 我们现有队列ADT 的一个实例
# newPrinTask 决定是否创建一个新的打印任务


import random

def simulation(numSeconds, pagesPerMinute):     # numSeconds 表示这段时间内
    
    labprinter = Printer(pagesPerMinute)
    printQueue = Queue()
    waitingtimes = []
    
    for currentSecond in range(numSeconds):
        
        if newPrintTask():
            """模拟排队场景 记录排队时刻"""
            task = Task(currentSecond)             
            printQueue.enqueue(task)
            
        if (not labprinter.busy()) and (not printQueue.isEmpty()):
            """模拟 打印机处理下一个任务  记录该任务的等待时间"""
            nexttask = printQueue.dequeue()
            waitingtimes.append(nexttask.waitTime(currentSecond))
            labprinter.startNext(nexttask)
            
        labprinter.tick()
        
    averageWait = sum(waitingtimes)/len(waitingtimes)
    print("Average Wait %6.2f secs %3d tasks remaining."%(averageWait, printQueue.size()))

# 模拟数据 建立在平均数上
# 假如一个小时内有10个学生 每人打印两次  一共就20个任务  也就是说 3600/20 = 180秒 可能会产生一个任务 
def newPrintTask():
    num = random.randrange(1, 181)
    if num == 180:
        return True
    else:
        return False
    
for i in range(10):
    simulation(3600, 5)
    
    
# 类 并不止代表 实体  
# 类 代表 一个结构体
```
